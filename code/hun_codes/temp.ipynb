```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(c,t):
    return 0.5*c*np.cos(2*np.pi*t/366.0)+(1-0.5*c)
```

```python
x = np.linspace(0,366, 366)
```

```python
plt.figure(figsize=(12,10))
for c in np.linspace(0,1,10):
    y = f(c, x)
    plt.plot(x,y, label=f"{c}")

plt.legend()
```

```python
y = f(0.5, x)
plt.plot(x,y, label=f"{c}")
```
